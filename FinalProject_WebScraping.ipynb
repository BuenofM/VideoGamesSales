#### Libraries

```python
import pandas as pd
import numpy 

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()

import requests as r
import urllib.request
from bs4 import BeautifulSoup

import time
import re
```

### A. First request: access the catalogue

```python
# Make request for first page of video games catalogue
url = 'https://www.instant-gaming.com/en/search/?all_types=1&all_cats=1&min_price=0&max_price=100&noprice=0&min_discount=0&max_discount=100&min_reviewsavg=10&max_reviewsavg=100&noreviews=0&available_in=FR&gametype=all&sort_by=&query='
response = r.get(url)
print(response.status_code)
```

```python
# Parse content of first page
soup=BeautifulSoup(response.content, 'html.parser')
soup
```

```python
# Get catalogue number of pages 
max_pages = soup.select('li a')[7].text
max_pages = int(max_pages)
max_pages
```

### B. Second set of requests:  all url's in catalogue

```python
# Make request for all pages of catalogue

step = 0
soups = []

for i in range(max_pages):
    url=f'https://www.instant-gaming.com/en/search/?all_types=1&all_cats=1&min_price=0&max_price=100&noprice=0&min_discount=0&max_discount=100&min_reviewsavg=10&max_reviewsavg=100&noreviews=0&available_in=FR&gametype=all&sort_by=&query=&page={i}'
    resp=r.get(url)
    if response.status_code == 200:
        soups.append(BeautifulSoup(resp.content, 'html.parser'))
        
        if step%5==0:
            time.sleep(2.4)
            print('page',step,'is finished')
        
    else:
        print('issue with page', step)
    
    step+=1
```

```python
# Get links for each game in catalogue

step = 1
links_games=[]

for i in soups:
    a = i.select('a.cover')
    for j in a:
        links_games.append(j.get('href'))
        
        if step%5==0:
            time.sleep(2)
            print(step, "url's have been saved")
        step+=1
```

### C. Access each URL and gather information on each Video Game

What information?
- Title `h1`
- Platform `a.platform`
- Console `div.platforms`
- Instant Download / In stock (yes/no) `div.download`
- Genres `div.tags`
- Number of langauges `div.languages`
- Retail Price `div.retail`
- Discount `div.discount`
- Release date `div.release span`
- User Rating `span.rate.rate9`
- Number of reviews `div.content h3`
- Number of comments `a.tab.feedbacks`

#### Accessing each url

```python
# Make request for each video-game page and parse it's content 
links_scraped = []
tomato_soup = []

for link in links_games:
    game_page = r.get(link).content
    game_soup = BeautifulSoup(game_page, 'html.parser')
        
    links_scraped.append(link)
    tomato_soup.append(game_soup)
    
    title = game_soup.select_one('h1')
    if type(title) == type(None):
        title = 'No information'
    else:
        title = title.get_text()
    print('Link for', title, 'is done.')
    print(link)
    
    time.sleep(1)
```

```python
if len(links_scraped) == len(links_games):
    print("All the url's were successfully scraped.")
else: 
    print("Ups. Not all url's were scraped!")
```

#### Scraping each url

```python
# Brouillon
for crouton in tomato_soup:
    title = crouton.select_one('h1').get_text()
    print(title)

    #Languages
    languages_info = crouton.select('div.languages')
    for i in languages_info:
        text = i.get_text().replace('\n', '').replace('Languages', '') 
        if len(text) >= 5:
            print('Languages:', text)
        else:
            #image = i.find('img')
            #langs = image.get('title')
            langs = i.find('img')['title']
            print('Languages:', langs)
        
    
    print('--')
```

```python
# For each video game in soup, scrape the information we need from the website 
#  and append it to respective lists so we can create a dataframe after

Titles = []
Platforms = [] 
Console = []
Edition = []
Stock_Availabilities = []
Download_Availabilities = []
Genres = []
Languages = []
Retail_Prices = []
Discounts = []
Release_Date = []
User_Ratings = []
Reviews = []
Comments = []


for crouton in tomato_soup:
    
    #Title
    title = crouton.select_one('h1').get_text()
    print('Title:', title)
    Titles.append(title)

    #Platform 
    platform = crouton.select_one('a.platform').get_text()
    print('Platform:', platform)
    Platforms.append(platform)

    #Console and Edition Specifications
    console_info = crouton.select_one('div.platforms div.button.edition')
    if type(console_info) == type(None):
        cons = 'No information'
        print('Console:', 'No information')
        Console.append(cons)
    else:
        cons = console_info.get_text().replace('\n', '')
        print('Console info:', cons)
        Console.append(cons)
        
    edition_info = crouton.select_one('div.editions div.button.edition')
    if type(edition_info) == type(None):
        edtn = 'No information'
        print('Edition:', 'No information')
        Edition.append(edtn)
    else:
        edtn = edition_info.get_text().replace('\n', '')
        print('Edition info:', edtn)
        Edition.append(edtn)

    #Download information
    download_info = crouton.select('div.download')
    if len(download_info) < 2:
        stock = download_info[0].get_text().replace('\n', '')
        download =  download_info[0].get_text().replace('\n', '')
        print('Stock:', stock)
        print('Download:', download)
        Stock_Availabilities.append(stock)
        Download_Availabilities.append(download)
    elif len(download_info) >= 2: 
        stock = download_info[0].get_text().replace('\n', '')
        download =  download_info[1].get_text().replace('\n', '')
        print('Stock:', stock)
        print('Download:', download)
        Stock_Availabilities.append(stock)
        Download_Availabilities.append(download)
    elif len(download_info) == 0:
        stock = 'No information'
        download = 'No information'
        print('Stock:', stock)
        print('Download:', download)
        Stock_Availabilities.append(stock)
        Download_Availabilities.append(download)

    #Genre
    genres = crouton.select_one('div.tags')
    if type(genres) == type(None):
        genres_list = 'No genre'
        print('Genre:', genres_list)
        Genres.append(genres_list)
    else:
        genres_list = genres.get_text().replace('\n', ',')
        genres_list = genres_list.lstrip(',').rstrip(',')
        print('Genre:', genres_list)
        Genres.append(genres_list)

    #Languages
    languages_info = crouton.select('div.languages')
    for i in languages_info:
        text = i.get_text().replace('\n', '').replace('Languages', '') 
        if len(text) >= 5:
            print('Languages:', text)
            Languages.append(text)
        else:
            #image = i.find('img')
            #langs = image.get('title')
            langs = i.find('img')['title']
            print('Languages:', langs)
            Languages.append(langs)

    #Retail price
    retail_price = crouton.select_one('div.retail')
    if type(retail_price) == type(None):
        retail_price = 'No price'
        print('Retail Price:', retail_price)
        Retail_Prices.append(retail_price)
    else:
        retail_price = retail_price.get_text().replace('\n', '').lstrip('Retail price:')
        print('Retail Price:', retail_price)
        Retail_Prices.append(retail_price)

    #Discount price
    discount = crouton.select_one('div.discount').get_text()
    print('Discount:', discount)
    Discounts.append(discount)

    #Release date
    date = crouton.select_one('div.release span').get_text()
    print('Release Date:', date)
    Release_Date.append(date)
    
    #User rating
    user_rating = crouton.select_one('div.tabs span.rate.rate9')
    if type(user_rating) == type(None):
        rating = 'No rating'
        print('User Rating:', rating)
        User_Ratings.append(rating)
    else:
        rating = user_rating.get_text()
        print('User Rating:', rating)
        User_Ratings.append(rating)

    #Total reviews
    total_reviews = crouton.select_one('div.content h3').get_text()
    total_reviews = re.sub("[^\d\.]", "", total_reviews)
    print('Total Reviews:', total_reviews)
    Reviews.append(total_reviews)

    #Total comments
    comments = crouton.select_one('a.tab.feedbacks')
    if type(comments) == type(None):
        total_comments = 'No comments'
        print('Total Comments:', total_comments)
    else:
        total_comments = comments.get_text().replace('\n', '')    
        total_comments = re.sub("[^\d\.]", "", total_comments)
        print('Total Comments:', total_comments)
        Comments.append(total_comments)
    
    print('-----------')
```

#### Saving everything into a dataframe

```python
ig_df = pd.DataFrame({'Titles': Titles,
     'Platforms': Platforms,
     'Consoles': Console,
     'Edition': Edition,
     'Stock_Availabilities': Stock_Availabilities,
     'Download_Availabilities': Download_Availabilities,
     'Genres': Genres,
     'Languages': Languages,
     'Retail_Prices': Retail_Prices,
     'Discounts': Discounts,
     'Release_Date': Release_Date,
     'User_Ratings': User_Ratings,
     'Reviews': Reviews,
     'Comments': Comments})

ig_df
```
